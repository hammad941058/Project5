```python
#dropping columns which has more than 50% null values
def remove_null_columns(df, threshold=0.5):
    """
    Remove columns that have more than a certain percentage of null values.

    Parameters:
    df (DataFrame): The input DataFrame.
    threshold (float): The threshold for removing columns (default is 0.5, i.e., 50%).

    Returns:
    DataFrame: The DataFrame with columns removed.
    """
    # Calculate the percentage of missing values in each column
    missing_percentage = df.isnull().mean()

    # Filter out columns where the missing percentage is greater than the threshold
    columns_to_keep = missing_percentage[missing_percentage <= threshold].index

    # Return the DataFrame with those columns
    return df[columns_to_keep]

# Usage example:
df_cleaned = remove_null_columns(df, threshold=0.5)
```

```python
def plot_outliers_for_numeric_columns(df):
    numeric_columns = df.select_dtypes(include=[np.number]).columns
    num_plots = len(numeric_columns) # Corrected variable name to num_plots
    plt.fig, axes = plt.subplots(nrows=num_plots, ncols=1, figsize=(8, 6 * num_plots)) # Corrected variable name to num_plots

    if num_plots == 1:
        # Indented the following lines to be part of the 'if' block
        sns.boxplot(data=df[numeric_columns[0]], ax=axes)
        axes.set_title(f'Boxplot for {numeric_columns[0]}')
    else:
        # Indented the following lines to be part of the 'else' block
        for i, column in enumerate(numeric_columns):
            sns.boxplot(data=df[column], ax=axes[i])
            axes[i].set_title(f'Boxplot for {column}')
        plt.tight_layout()
        plt.show()
plot_outliers_for_numeric_columns(df)
```

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df[(df[column]>=lower_bound) & (df[column]<=upper_bound)]
```

```python
# Calculate total data usage for each user by summing download and upload data
user_data_usage = df_cleaned.groupby('MSISDN/Number').agg({
    'Total DL (Bytes)': 'sum',
    'Total UL (Bytes)': 'sum'
})

# Calculate the total data volume per user
user_data_usage['Total Data Usage (Bytes)'] = user_data_usage['Total DL (Bytes)'] + user_data_usage['Total UL (Bytes)']

# Determine quantiles for segmentation
low_threshold = user_data_usage['Total Data Usage (Bytes)'].quantile(0.25)  # Bottom 25%
high_threshold = user_data_usage['Total Data Usage (Bytes)'].quantile(0.75) # Top 25%

# Create a function to label data consumers
def classify_data_usage(total_data):
    if total_data >= high_threshold:
        return 'High'
    elif total_data >= low_threshold:
        return 'Medium'
    else:
        return 'Low'

# Apply classification to each user
user_data_usage['Data Consumer Segment'] = user_data_usage['Total Data Usage (Bytes)'].apply(classify_data_usage)

# Display the first few rows to check results
print(user_data_usage.head())
```

```python
# Define least engaged baseline as the minimum values of engagement features
least_engaged_point = np.array([
    user_engagement['Total Data Usage (Bytes)'].min(),
    user_engagement['Number_of_Sessions'].min(),
    user_engagement['Average_Session_Duration (ms)'].min()
])

# Define worst experience baseline as the minimum values of experience features
worst_experience_point = np.array([
    user_engagement['Average_Session_Duration (ms)'].min(),
    user_engagement['Social_Media_Usage_Ratio'].min(),
    user_engagement['YouTube_Usage_Ratio'].min(),
    user_engagement['Netflix_Usage_Ratio'].min(),
    user_engagement['Gaming_Usage_Ratio'].min()
])
```

```python
from scipy.spatial.distance import euclidean

# Calculate Engagement Score as Euclidean distance to the least engaged point
user_engagement['Engagement_Score'] = user_engagement.apply(
    lambda row: euclidean(
        [row['Total Data Usage (Bytes)'], row['Number_of_Sessions'], row['Average_Session_Duration (ms)']],
        least_engaged_point
    ),
    axis=1
)

# Calculate Experience Score as Euclidean distance to the worst experience point
user_engagement['Experience_Score'] = user_engagement.apply(
    lambda row: euclidean(
        [row['Average_Session_Duration (ms)'], row['Social_Media_Usage_Ratio'], row['YouTube_Usage_Ratio'],
         row['Netflix_Usage_Ratio'], row['Gaming_Usage_Ratio']],
        worst_experience_point
    ),
    axis=1
)

# Display the calculated scores
print("Engagement and Experience Scores:\n", user_engagement[['Engagement_Score', 'Experience_Score']].head())
```

```python
# Select Engagement and Experience Scores for clustering
clustering_data = user_engagement[['Engagement_Score', 'Experience_Score']]
```

```python
from sklearn.cluster import KMeans

# Calculate WCSS for different numbers of clusters
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(clustering_data)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='-')
plt.title('Elbow Method for Optimal Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.show()
```

```python
# Assuming the optimal number of clusters is 3 (based on the Elbow Method)
optimal_clusters = 3

# Apply KMeans clustering
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
user_engagement['Cluster'] = kmeans.fit_predict(clustering_data)

# Display the first few rows with cluster labels
print(user_engagement[['Engagement_Score', 'Experience_Score', 'Cluster']].head())
```
